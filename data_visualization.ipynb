```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("ds-all-data.csv")
df["InterestRate"].plot.bar(figsize=(50,10), fontsize=55)
# this is a bad data visualization, please never do this.
```

```python
interest_rate = df["InterestRate"].copy()
interest_rate.sort_values(inplace=True)
interest_rate.plot.bar(figsize=(50,10), fontsize=55)
```

```python
value_counts = interest_rate.value_counts()
value_counts.plot.bar(figsize=(50,10), fontsize=55, title="Interest Rate Frequency Count")
font = {'size':55}
matplotlib.rc('font', **font)
```

```python
value_counts.plot.pie(figsize=(7,7), fontsize=55, title="Interest Rate Frequency Count")
font = {'size':35}
matplotlib.rc('font', **font)
```

```python
value_counts.plot.line(figsize=(7,7), fontsize=25, title="Interest Rate Frequency Count")
font = {'size':15}
matplotlib.rc('font', **font)
```

```python
age = df["Age"].copy()
#age.sort_values(inplace=True)
#age.plot()
age = age.value_counts()
ages = list(age.index)
frequencies = [age.ix[i] for i in age.index]

plt.plot(frequencies, ages)
plt.xlabel('Frequencies')
plt.ylabel('Ages')
plt.axis()
plt.show()
```
